```python
#Se importa el archivo y se obtiene su forma. 
#Método basado en el algoritmo descrito en el repositorio de María Camila Remolina
#Datos en archivo de entrada organizados de la siguiente manera:
#CODIGO - MUNICIPIO - DEPARTAMENTO - HOMICIDIOS_AÑO - POBLACION_AÑO - TASA_AÑO

with open('homicidios.1990.a.2013.csv','rb') as a:
    archivo = csv.reader(a)
    filas=0
    columnas=0
    k=0
    for f in archivo:
        if(k==0):
            for c in f:
                columnas+=1
        filas+=1
        k+=1
print filas
print columnas
```

```python
#Se guardan los datos en una matriz que se pueda manejar fácilmente
with open('homicidios.1990.a.2013.csv','rb') as a:
    archivo = csv.reader(a)
    #Número de dimensiones= número de columnas menos 3 puesto que las primeras tres columnas son informativas
    dimensiones=(columnas-3)
    #Número de elementos= número de filas menos 1 puesto que la primera fila indica lo que contiene cada columna
    N=filas-1
    #Arreglos y matriz con datos organizados
    codigo_dato=[]
    municipio=[]
    departamento=[]
    datos=zeros((N,dimensiones)) 
    i=0
    for f in archivo:
        if(i!=0):         
            j=0
            for c in f:
                if(j==0):
                    #Guardo código del municipio
                    codigo_dato.append(c)
                elif(j==1):
                    #Guardo nombre del municipio
                    municipio.append(c)
                elif(j==2):
                    #Guardo nombre del departamento
                    departamento.append(c)
                else:
                    #Guardo datos
                    if (c=='NA' or c=='#DIV/0!'):
                        datos[i-1,j-3]=0
                    else:
                        datos[i-1,j-3]=c
                j+=1
        i+=1
print dimensiones
print N
```

```python
#Se arregla la matriz para que tenga únicamente los valores de tasas de homicidio por año,
#reduciendo así las dimensiones a un tercio quitando la información redundante
numero_anios=dimensiones/3
tasas=zeros((N,numero_anios))
for i in range(N):
    for j in range(numero_anios):
        tasas[i,j]=datos[i,(j*3)+2]
```

```python
#Se halla el valor medio de cada tasa de homicidios por año
medios=zeros(numero_anios)

for i in range(numero_anios):
    medios[i]=mean(tasas[:,i])
```

```python
#Se halla la matriz de covarianza
numerador = zeros(N)
matriz_cov = zeros((numero_anios,numero_anios))

for i in range(numero_anios):
    for j in range(numero_anios):
        for k in range(N):
            numerador[k]=(tasas[k,i]-medios[i])*(tasas[k,j]-medios[j])
        nume_total=sum(numerador)
        valor=nume_total/(N-1)
        matriz_cov[i,j]=valor
        
imshow(matriz_cov)
```

```python
#Autovalores y autovectores correspondientes
autovalores, autovectores = linalg.eig(matriz_cov)

numero_autoval=zeros(size(autovalores))
suma_autovalores=zeros(size(autovalores))
#Se ordenan de manera descendente los autovalores
nuevos_autoval=-sort(-autovalores)

#Criterio de componentes principales significativos
significancia=0.95
#Se halla el acumulado de los autovalores y se grafica contra el número de autovalor
for i in range(size(autovalores)):
    numero_autoval[i]=i+1
    suma_autovalores[i]=suma_autovalores[i-1]+nuevos_autoval[i]
total=suma_autovalores[size(autovalores)-1]
acumulado=suma_autovalores/total
linea=significancia+zeros(size(autovalores))

scatter(numero_autoval,acumulado)
plot(numero_autoval,linea,c='r')
xlabel("Numero de componente principal")
ylabel("Acumulado")
axis([0,25,0,1.1])

#Se extrae el número de componentes principales más relevantes (según la significancia) a partir de los cuales se pueden expresar
#los datos como una combinación lineal de ellos
componentes_principales=0
for i in range(size(autovalores)):
    if(acumulado[i]<significancia):
        componentes_principales=i+1
print "Los primeros",componentes_principales,"componentes principales abarcan el",significancia*100,"% del valor total, por lo que el problema se puede dimensionar a este número"
```
